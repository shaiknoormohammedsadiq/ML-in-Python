```python
sns.relplot(x='total_bill',y='tip',data=data,hue="smoker",size="day",col="time",style='sex')
```

```python
sns.lmplot(x="total_bill", y="tip", col="time", hue="smoker",data=data);
```

```python
import pandas as pd
import numpy as np
d1 = {'Names':['Abdul','Raju','Siva','Sunil','Aravind'],
     'Rollnos':np.arange(501,506),
     'Python':np.random.randint(0,100,5),
     'Java':np.random.randint(0,100,5),
     'C':np.random.randint(0,100,5),
     'CPP':np.random.randint(0,100,5),
     'Android':np.random.randint(0,100,5),
     'PHP':np.random.randint(0,100,5)}
data=pd.DataFrame(d1)
data
```

```python
data.to_csv('marks')
```

```python
marks_data = pd.read_csv('marks')
marks_data=marks_data.drop(['Unnamed: 0'],axis=1)
```

```python
marks_data
```

```python
li=[]
li.append(marks_data['Python'].max())
li.append(marks_data['Java'].max())
li.append(marks_data['C'].max())
li.append(marks_data['CPP'].max())
li.append(marks_data['Android'].max())
li.append(marks_data['PHP'].max())
li
```

```python
cols=data.columns[2:]
```

```python
cols
```

```python
import matplotlib.pyplot as plt
plt.bar(cols,list(li),color='rgb')
```

```python
plt.pie(li,labels=cols,autopct='%.1f%%')
```
