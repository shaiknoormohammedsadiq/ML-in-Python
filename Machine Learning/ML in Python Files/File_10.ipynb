```python
# i want draw more than one plot
x1 = np.array([1,2,3,4,5])
y1 = np.array([5,2,9,7,1])
x2 = x1*2
y2 = y1*2
x1,y1,x2,y2
```

```python
plt.plot(x1,y1,'g-',x2,y2,'b-')
```

```python
plt.plot(x1,y1,'g-',x2,y2,'b-')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('Sample Practice')
plt.show()
```

```python
plt.plot(x1,y1,'g-o',label='line 1')
plt.plot(x2,y2,'b-^',label='line 2')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend()# without calling legend() we can't show plot labels
plt.title('Sample Practice')
plt.show()
```
