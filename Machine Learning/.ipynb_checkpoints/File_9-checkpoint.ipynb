## merging

```python
stu_data = pd.DataFrame({'Name':['Gokul','Srikanth','Sai Kiran'],
                        'Rool Nos':['5C5','502','539'],
                        'Branch':['CSE','CSE','CSE']})
stu_data
```

```python
valid_info=pd.DataFrame({'Name':['Gokul','Upendra','Sai Kiran'],
                       'Love Status':['Yes','Yes','Yes'],
                       'Lovers':[2,2,3]})
valid_info
```

```python
pd.merge?
```

```python
pd.merge(stu_data,valid_info)
```

```python
pd.merge(stu_data,valid_info,how='left')
```

```python
pd.merge(stu_data,valid_info,how='right')
```

```python
pd.merge(stu_data,valid_info,how='outer')
```

```python
valid_info['Rool Nos']=['5C5','532','539']
```

```python
valid_info
```

```python
pd.merge(stu_data,valid_info)
```

```python
pd.merge(stu_data,valid_info,on='Rool Nos')
```

```python
df = pd.merge(stu_data,valid_info,on='Rool Nos',how='outer')
```

```python
df
```

```python
df.isnull().sum()
```

```python
df.dropna()
```

```python
df
```

```python
df.dropna(inplace=True)
```

```python
df
```

```python
df = pd.merge(stu_data,valid_info,on='Rool Nos',how='outer')
```

```python
df
```

```python
df.isnull().sum()
```
